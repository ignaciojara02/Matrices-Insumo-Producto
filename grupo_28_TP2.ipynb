$$lim_{n‚Üí‚àû} A^{n+1} = 0.$$

Esto asegura que, al elevar $A$ a potencias sucesivas, el valor de $A^{n}$ se reduce progresivamente, hasta que $A^{n}$ tiende a la matriz nula cuando $n ‚Üí ‚àû$.

 $$ S_n = ‚àë_{k=0}^{n} A^{k} = I + A + A^{2} + ‚ãØ + A^{n}. $$

 a) Observemos qu√© sucede anal√≠ticamente al multiplicar la matriz $E_n$ por un vector gen√©rico $z ‚àà \mathbb{R}^{n} = \begin{pmatrix} z_1 \\ z_2 \\ \vdots \\ z_n \end{pmatrix}$.

  Notamos que la matriz $E_n$ tiene la forma $$\begin{pmatrix} (1-1/n) & -1/n & ‚ãØ & -1/n \\ -1/n & (1-1/n) & ‚ãØ & -1/n \\ \vdots & \vdots&\ddots  & \vdots \\ -1/n & -1/n & \cdots & (1-1/n) \end{pmatrix}$$ y al hacer el producto por un vector columna, obtenemos un vector cuyas componentes resultan ser la suma de las filas de la matriz.
  Siendo
  
  $$E_nz = \begin{pmatrix} (1-1/n)z_1-1/nz_2-\cdots-1/nz_n \\ -1/nz_1+(1-1/n)z_2-\cdots-1/nz_n\\\vdots \\ -1/nz_1-1/nz_2 -\cdots-(1-1/n)z_n\end{pmatrix}$$
  $$ E_n z = (I_n - \frac{1}{n} ee^T)z = z - \frac{1}{n} e \sum_{i=1}^{n} z_i. $$

  Entonces, cada componente de $E_n z$ es:

  $$ (E_n z)_i = z_i - \frac{1}{n} \sum_{j=1}^{n} z_j.$$

  Esto implica que $E_n z$ es el vector z, al que se le resta el promedio de sus componentes de manera uniforme.
  El promedio de las componentes de $E_n z$ se calcula como:

  $$\frac{1}{n} \sum_{i=1}^{n} (E_n z)_i = \frac{1}{n} \sum_{i=1}^{n} \left( z_i - \frac{1}{n} \sum_{j=1}^{n} z_j \right).$$

  Distribuyendo la suma:

  $$\frac{1}{n} \sum_{i=1}^{n} z_i - \frac{1}{n} \cdot \frac{1}{n} \sum_{i=1}^{n} \sum_{j=1}^{n} z_j.$$

  El primer t√©rmino es el promedio de $z$, que llamamos
  $\overline{z}$:

  $$\frac{1}{n} \sum_{i=1}^{n} z_i = \overline{z}.$$

  El segundo t√©rmino se simplifica porque $ \sum_{i=1}^{n} \sum_{j=1}^{n} z_j = n \sum_{j=1}^{n} z_j.$ Entonces:

  $$\frac{1}{n} \cdot \frac{1}{n} n \sum_{j=1}^{n} z_j = \frac{1}{n} \sum_{j=1}^{n} z_j = \overline{z}.$$

  Por lo tanto:

  $$\frac{1}{n} \sum_{i=1}^{n} (E_n z)_i = \overline{z} - \overline{z} = 0.$$

  Por lo que $E_n z$ es 0 para cualquier $ z \in \mathbb{R}^n $

  $$E_n A = \left[ E_n a_1, E_n a_2, \dots, E_n a_n \right]$$

  Multiplicar un vector por $E_n$ sustrae el promedio de sus componentes
  Aplicando esto a cada columna $a_i$, tenemos:

  $$E_n a_i = a_i - \frac{1}{n} e \sum_{j=1}^{n} (a_i)_j = a_i - \overline{a}_i e$$

  $$ E_n A = \left[ (a_1 - \overline{a}_1 e)   (a_2 - \overline{a}_2 e) \dots (a_n - \overline{a}_n e) \right] $$


 Multiplicar una matriz $A$ por $E_n$ transforma cada columna $a_i$en una nueva columna cuyo promedio es 0. Por lo tanto, $E_n$ elimina la componente promedio de cada columna de $A$.

(I) $E_n$ es matriz ortogonal. Para que $E_n$ sea una matriz ortogonal, debe cumplirse que $E_n^tE_n=I_n$ pero como vimos que $E_n$ tiene un autovalor $Œª=0$, esto implica que no tiene inversa, por lo que la afirmaci√≥n es falsa.

(II) $E_n$ es matriz singular. En relaci√≥n al punto anterior, tomando que $E_n$ tiene un autovalor 0 podemos afirmar que es singular. La afirmaci√≥n es verdadera.

(III) $E_n$ es definida positiva. Para que $E_n$ sea definida positiva debe ocurrir que para todos los vectores $v\in \mathbb{R}^n$ y $v \neq 0$ se cumple que: $vE_nv^t>0$, lo cual es falso puesto que al $E_n$ ser una matriz singular, existe un vector $x\in \mathbb{R}^n$ y $x \neq 0$ que hace $E_vx=0$, en particular podemos afirmar que existe un vector $x$ que hace $x^tE_nx=0$ por lo que aseguramos que $E_n$ no es definida positiva. Es falsa.

(IV) $Traza(E_n) = n-1$. Hab√≠amos observado que la forma matricial de $E_n$ solo tiene $(1- \frac{1}{n})$ en la diagonal. La traza se define como la suma de los elementos de la diagonal, en resumen: $Traza(E_n) = n(1-\frac{1}{n})=n-1$ por lo que la afirmaci√≥n es verdadera.

Para relacionar la matriz $H=A(I-A)^{-1}$ con la suma de potencias debo considerar la expansi√≥n en serie infinita de la inversa de $I-A$ vista en la consigna 1 donde: $$(I-A)^{-1} = I+A+A^2+A^3\dots$$
Luego, sustituyo esta expresi√≥n en $H$:
$$H=A(I-A)^{-1}=A(I+A+A^2+A^3\dots).$$
Distribuyendo observo que puedo escribir a $H$ como:
$$H=A+A^2+A^3+‚Ä¶$$
Entonces la relaci√≥n entre $H$ y la suma de potencias es que es posible escribirla como una suma de potencias de $A$ siempre que esta serie converja. Esta suma se puede interpretar como todas las interacciones de orden $n$ entre sectores representados por $A$. Esto asocia a la matriz $H$ con un refuerzo a los sectores que ya tienen fuertes relaciones en $A$ ya que suma todas relaciones directas e indirectas entre sectores.
